```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv(r"C:\Users\Purva\Documents\CL-1\5-Sales & Customer\customer_shopping_data.csv")
df
```

```python
df.head()
```

```python
df.groupby("shopping_mall").count()
```

```python
df.groupby("category").count()
```

```python
branch_sales = df.groupby("shopping_mall").sum()
```

```python
category_sales = df.groupby("category").sum()
```

```python
branch_sales.sort_values(by = "price", ascending = False)
```

```python
category_sales.sort_values(by = "price", ascending = False)
```

```python
plt.figure(figsize=(6, 6))  # use plt.figure instead of plt.Figure
plt.pie(
    branch_sales["price"],
    labels=branch_sales.index,
    autopct="%1.1f%%",   # shows percentage with 1 decimal place
    startangle=90
)
plt.title("Sales Distribution by Branch")
plt.show()
```

```python
plt.figure(figsize=(6, 6))  # use plt.figure instead of plt.Figure
plt.pie(
    category_sales["price"],
    labels=category_sales.index,
    autopct="%1.1f%%",   # shows percentage with 1 decimal place
    startangle=90
)
plt.title("Sales Distribution by Branch")
plt.show()
```

```python
combined_pivot = df.pivot_table(index="shopping_mall", columns="category", values="price", aggfunc="sum")
# grouped bar chart for sales of different categories at different branches
combined_pivot.plot(kind="bar", figsize=(10, 6))
plt.show()
```
